```python
nasa = pd.read_csv('ExoplanetData.csv')
data = nasa[nasa['default_flag'] == 1]
data2 = pd.read_csv('open_exoplanet_catalogue.txt')
```

```python
print(len(data))
# Insolation flux between 0.3 and 1.5
data['habitable_zone'] = data['pl_insol'].apply(
    lambda flux: 'Yes' if 0.3 <= flux <= 1.5 else 'No' if pd.notna(flux) else None
)
data_stage1 = data[data['habitable_zone'] == 'Yes']
data_stage1.to_csv("candidates_after_criteria_1.csv", index=False)

# Planetary temperature between 198 and 320 K
data.loc[data['habitable_zone'] == 'Yes', 'habitable'] = data.loc[data['habitable_zone'] == 'Yes'].apply(
    lambda row: 'Yes' if pd.notna(row['pl_eqt']) and 198 <= row['pl_eqt'] <= 320 else 'No',
    axis=1
)
data_stage2 = data[data['habitable'] == 'Yes']
data_stage2.to_csv("candidates_after_criteria_2.csv", index=False)

# Planetary radius between 0.5 and 2.0 Earth radii
data.loc[data['habitable_zone'] == 'Yes', 'habitable_zone'] = data.loc[data['habitable_zone'] == 'Yes'].apply(
    lambda row: 'Yes' if pd.notna(row['pl_rade']) and 0.5 <= row['pl_rade'] <= 2.0 else 'No',
    axis=1
)
data_stage3 = data[data['habitable_zone'] == 'Yes']
data_stage3.to_csv("candidates_after_criteria_3.csv", index=False)

# Planetary mass between 0.1 and 10 Earth masses
data.loc[data['habitable_zone'] == 'Yes', 'habitable_zone'] = data.loc[data['habitable_zone'] == 'Yes'].apply(
    lambda row: 'Yes' if pd.notna(row['pl_bmasse']) and 0.1 <= row['pl_bmasse'] <= 10 else 'No',
    axis=1
)
data_stage4 = data[data['habitable_zone'] == 'Yes']
data_stage4.to_csv("candidates_after_criteria_4.csv", index=False)

# Planetary mass between 0.1 and 10 Earth masses (for stars with spectral types G, K, M)
data.loc[data['habitable_zone'] == 'Yes', 'habitable_zone'] = data.loc[data['habitable_zone'] == 'Yes'].apply(
    lambda row: 'Yes' if pd.notna(row['st_spectype']) and str(row['st_spectype']).strip()[0] in ['G', 'K', 'M'] else 'No',
    axis=1
)
data_stage5 = data[data['habitable_zone'] == 'Yes']
data_stage5.to_csv("candidates_after_criteria_5.csv", index=False)


```

```python
# Define a color palette for broad spectral types
spectral_palette = {
    'O': 'blue', 'B': 'royalblue', 'A': 'cyan', 'F': 'green', 
    'G': 'yellow', 'K': 'orange', 'M': 'red', 'L': 'purple',
    'T': 'brown', 'Y': 'pink', 'm': 'white', 'D': 'gray'  # White Dwarfs (D)
}

# Select two properties for the scatter plot (e.g., Planet Radius vs. Orbital Period)
x_col = "pl_rade"  # Planet Radius in Earth radii
y_col = "pl_orbper"  # Orbital Period in days

# Filter out NaNs
df_filtered = data[[x_col, y_col, "st_spectype"]].dropna()

df_filtered['spectype_2'] = df_filtered['st_spectype'].str[0]

# Create scatter plot
plt.figure(figsize=(10, 6))
sns.scatterplot(
    data=df_filtered, 
    x=x_col, 
    y=y_col, 
    hue=df_filtered["spectype_2"], 
    palette=spectral_palette,
    alpha=0.7  # Transparency for better visibility
)

plt.xscale("log")  # Optional: Log scale if values span large ranges
plt.yscale("log")
plt.xlabel("Planet Radius (Earth Radii)")
plt.ylabel("Orbital Period (days)")
plt.title("Exoplanets: Orbital Period vs. Planet Radius, Colored by Spectral Type")
plt.legend(title="Spectral Type", bbox_to_anchor=(1.05, 1), loc='upper left')
plt.grid(True, which="both", linestyle="--", alpha=0.5)
plt.show()
```
